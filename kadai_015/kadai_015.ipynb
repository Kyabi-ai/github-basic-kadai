```python
#コントラスト化の奴
class Printinfo:
  def __init__(self,name,age):
    self.name = name
    self.age = age

printinfo = Printinfo("きゃびー",459)

print(printinfo.name)
print(printinfo.age)
```

```python

class Human:
    def __init__(self):
        self.name = ""
        self.aga = ""

    # メソッドを定義する
    #名前
    def set_name(self, name):
        self.name = name
    def show_name(self):
        print(self.name)
    #年齢
    def set_aga(self, aga):
      self.aga = aga
    def show_aga(self):
      print(self.aga)

# インスタンス化する
human = Human()

# メソッドにアクセスして実行する
human.set_name("きゃびー")
human.show_name()
human.set_aga(495)
human.show_aga()
```
