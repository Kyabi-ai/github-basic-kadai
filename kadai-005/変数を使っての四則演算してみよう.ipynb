```python
# inputだけだと文字データ（str）となり計算してくれないため、前に浮動小数点（float）に変換し数字のデータとして計算してもらう。
#整数型（int)にすると、小数点の計算が行われないため、今回はfloatにした。

top_side = float(input("上底を入力: "))
bottom_side = float(input("下底を入力: "))
height = float(input("高さを入力: "))

area=(top_side + bottom_side) * height / 2

print(str(area)+"㎠")
```

```python
top_side = int(input("上底を入力: "))
bottom_side = int(input("下底を入力: "))
height = int(input("高さを入力: "))

area=(top_side + bottom_side) * height / 2

print(str(area)+"㎠")
```

```python
def get_int(message):
    while True:
        try:
            return int(input(message))
        except ValueError:
            print("整数お願いだっピ★")

top_side = get_int("上底を入力: ")
bottom_side = get_int("下底を入力: ")
height = get_int("高さを入力: ")

area = (top_side + bottom_side) * height / 2

print(f"台形の面積は {area} ㎠ です")
```
